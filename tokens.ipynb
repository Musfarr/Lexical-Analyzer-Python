```python

import re

keywords = [
    "let", "var", "const", "read", "return","else", "if", "switch",
    "break","new", "int", "float", "char", "do", "string", "null","create"
    ,"cycle","true","false","empty","class","self","abstract","sealed","for"
]

punctuators = ['(', ')', '.', '/', ':', ';', '?', '[', ']', '{', '}', '~', '\"','/!/',',','']

operators = ['==', '+', '-', '*', '/', '%', '=', '<', '>', '<=', '>=', '!', '&&', '||', '!=', '++', '+=', '--', '-=', '?:', '<<', '>>', '!==', '^', '**']

def is_number(num):
    return re.match(r'^\s*-?\d+(\.\d+)?\s*$', num)

def tokenizer(input_string):
    tokens = []
    line = 1
    index = 0
    length = len(input_string)
    
    while index < length:
        char = input_string[index]
        
        if char in [' ', '\t', '\n']:
            index += 1
            if char == '\n':
                line += 1
            continue
        
        if re.match(r'[a-zA-Z_]', char):
            identifier = ''
            while char and re.match(r'[a-zA-Z0-9_]', char):
                identifier += char
                index += 1
                if index < length:
                    char = input_string[index]
                else:
                    char = None
            if identifier in keywords:
                tokens.append(['Keyword', identifier, line])
            elif re.match(r'^[_a-zA-Z][_a-zA-Z0-9]*$', identifier):
                tokens.append(['Identifier', identifier, line])
            else:
                tokens.append(['Invalid identifier', identifier, line])
            continue
        
        if re.match(r'[0-9]', char):
            number = ''
            while char and re.match(r'[0-9.]', char):
                number += char
                index += 1
                if index < length:
                    char = input_string[index]
                else:
                    char = None
            
            if is_number(number):
                tokens.append(['Number', number, line])
            continue
        
        if char in punctuators:
            tokens.append(['Punctuator', char, line])
            index += 1
            continue
        
        operator = ''
        while char and char not in punctuators and not re.match(r'[\w\s]', char):
            operator += char
            index += 1
            if index < length:
                char = input_string[index]
            else:
                char = None
        
        if operator in operators:
            tokens.append(['Operator', operator, line])
        else:
            tokens.append(['Invalid operator', operator, line])
    
    return tokens



input_string = "int a=7*g/b.a.r==y.u * 9"

#input_string="a[[1,2,3,4,5,6,7]]"


'''
input_string="""for int i=0; i<5; ++:
print(i)
/!/
"""
'''


'''
input_string="""for int i=0; i<5; ++:
a=0
if(a>9):
print(not_valid)
/!/
/!/
"""
'''




tokens = tokenizer(input_string)

# Write the tokens to a file
with open("output.txt", "w") as f:
    for token in tokens:
        f.write(str(token) + "\n")

print("Output written to output.txt")
 
```
